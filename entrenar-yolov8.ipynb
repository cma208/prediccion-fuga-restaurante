```python
from ultralytics import YOLO
from dotenv import load_dotenv
import os

# Cargar variables del archivo .env
load_dotenv()
data_path = os.getenv("DATA_PATH")

if not data_path:
    raise ValueError("No se encontrÃ³ la variable DATA_PATH en el archivo .env")

# Cargar el modelo base
model = YOLO("yolov8n.pt")

# Lista de batch sizes a probar (de mayor a menor)
for b in [32, 24, 16, 8, 4]:
    try:
        print(f"\nðŸ§ª Probando batch size: {b}")
        model.train(
            data=data_path,
            epochs=1,       # entrena solo una Ã©poca
            imgsz=640,      # respetamos tamaÃ±o de Roboflow
            batch=b,
            device="mps"
        )
        print(f"âœ… Batch size {b} funcionÃ³ correctamente.")
        break
    except Exception as e:
        print(f"âŒ Batch size {b} fallÃ³:\n{e}")
```

```python
# Entrenar modelo YOLOv8
from ultralytics import YOLO
from dotenv import load_dotenv
import os

# Cargar variables del archivo .env
load_dotenv()
data_path = os.getenv("DATA_PATH")

model = YOLO("yolov8n.pt")  # o yolov8s.pt, yolov8m.pt, etc.

model.train(
    data=data_path,
    epochs=50,
    imgsz=640,
    batch=32,
    device="mps"
)
```
